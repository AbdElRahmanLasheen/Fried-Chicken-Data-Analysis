## I. Data Acquisistion

```python
#Importing Libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from googletrans import Translator
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
#Reading Dataset Using Pandas
fc = pd.read_csv('Fried Chicken.csv')
fc
```

```python
#A lot of Text is in Arabic, so We Use Google Translate Library to Translate into English for analysis
translator = Translator()
```

```python
#Translation Function
def translate_text(text):
    if pd.isnull(text):  # Skip NaN values
        return text
    try:
        translated = translator.translate(text, src='ar', dest='en')
        return translated.text
    except Exception as e:
        print(f"Error translating: {text} - {e}")
        return text  # Return original if error occurs
```

```python
#Columns That Will be Translated
columns_to_translate = ['Choice Changes','Commercial Impacts']
```

```python
for column in columns_to_translate:
    fc[column + '_English'] = fc[column].apply(translate_text)
fc.to_csv("Fried Chicken Translated.csv", index=False)
```

## III. Data Analysis

```python
#Using Vader Sentiment for Sentiment Analysis on Choice Changes Column
analyzer = SentimentIntensityAnalyzer()
```

```python
def categorize_sentiment(text):
    sentiment = analyzer.polarity_scores(text)
    compound_score = sentiment['compound']
    
    # Categorize based on sentiment scores
    if compound_score >= 0.5:
        return 'High Positive Impact'
    elif 0.1 <= compound_score < 0.5:
        return 'Medium Positive Impact'
    elif -0.1 < compound_score < 0.1:
        return 'Neutral Impact'
    elif -0.5 < compound_score <= -0.1:
        return 'Medium Negative Impact'
    else:
        return 'High Negative Impact'
```

```python
cf['Sentiment Category'] = cf['Choice Changes'].apply(categorize_sentiment)
cf
```

```python
cf = cf.drop(['Choice Changes'] ,axis=1)
cf.rename(columns={'Sentiment Category': 'Choice Impact'}, inplace=True)
```
